```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
x = df[["Age","Fare"]]
y = df["Survived"]
```

```python
x
```

```python
y
```

```python
plt.scatter(x["Fare"],x["Age"],c=y)
plt.xlabel("Fare")
plt.ylabel("Age")
plt.show()
```

```python
plt.scatter(x["Fare"],x["Age"],c=y)
plt.plot([30,20],[0,50])
plt.xlabel("Fare")
plt.ylabel("Age")
plt.show()
```
